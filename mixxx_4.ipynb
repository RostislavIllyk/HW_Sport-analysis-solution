```python
import numpy as np
import pandas as pd
```

```python
base         = pd.read_csv('./result/result_zero.csv')
profile_part = pd.read_csv('./result/result_profile_part.csv')
applic_part  = pd.read_csv('./result/result_applic_part.csv')
bki_part  = pd.read_csv('./result/result_bki_part.csv')


base = base.set_index('APPLICATION_NUMBER')
profile_part = profile_part.set_index('APPLICATION_NUMBER')
applic_part = applic_part.set_index('APPLICATION_NUMBER')
bki_part = bki_part.set_index('APPLICATION_NUMBER')
```

```python
base = base.rename({'TARGET':'TARGET_profile'}, axis=1)
applic_part = applic_part.rename({'TARGET':'TARGET_app'}, axis=1)
bki_part = bki_part.rename({'TARGET':'TARGET_bki'}, axis=1)
```

```python
result=pd.concat([base, profile_part], axis=1)
result=pd.concat([result, applic_part], axis=1)
result=pd.concat([result, bki_part], axis=1)
result['TARGET_bki_applic_mul'] = (result['TARGET_app'].values * result['TARGET_bki'].values)**(1/2)
result
```

```python
result.loc[result['TARGET'].isnull(), 'TARGET'] = result['TARGET_bki_applic_mul']
result.loc[result['TARGET'].isnull(), 'TARGET'] = result['TARGET_bki']
result.loc[result['TARGET'].isnull(), 'TARGET'] = result['TARGET_app']
result.loc[result['TARGET'].isnull(), 'TARGET'] = result['TARGET_profile']


result = result.drop(['TARGET_profile','TARGET_app','TARGET_bki','TARGET_bki_applic_mul'], axis=1)
```
